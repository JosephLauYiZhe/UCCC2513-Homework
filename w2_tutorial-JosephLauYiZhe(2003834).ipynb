#### List and tuple
List and tuple are both variables that can store multiple items. The major differences between the 2 are:
* Elements in list is enclosed with square brackets, while tuple is written with parentheses.
* List is changeable, while tuple is unchangeable.

````python
mylist = ["a", "b", "c"]
print(mylist)

mytuple = ("a", "b", "c")
print(mytuple)
````

````python
mylist[0] = 1
print(mylist)
````

````python
try:
    mytuple[0] = 1
except Exception as e:
    print("Tuple is immutable")
````

## Custom function

````python
def simple_add(x, y):
    # It is a good practice to include docstring
    """Function to add 2 numbers:
    Parameters:
    x & y: must be numbers
    
    Returns:
    result of addition"""
    return (x + y)
````

````python
simple_add(2, 10)
````

But the above function is not robust, since it will invoke error if user provide non-numeric arguments.

````python
import numbers

def simple_add(x, y):
    """Function to add 2 numbers:
    Parameters:
    x & y: must be numbers
    
    Returns:
    result of addition"""

    if isinstance(x, numbers.Number) and isinstance(y, numbers.Number):
        return x + y
    else:
        print("Invalid input argument")
````

````python
simple_add(20, -10)
````

````python
simple_add(7, "a")
````

Another way to control Exception: use `try` and `except` blocks

````python
def simple_add(x, y):
    """Function to add 2 numbers:
    Parameters:
    x & y: must be numbers
    
    Returns:
    result of addition"""
    try:
        return x + y
    except Exception as e:
        print(e)
````

````python
simple_add(7, "a")
````

## Numpy module
Numpy is a Python module used for working with arrays. Numpy provides array object that is faster than Python lists. This is due to locality of reference. NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently. 

````python
import numpy as np
````

## Custom image using Numpy
Now that you have some basic understanding of the data structure of digital image, lets try to create a custom image with the help of Numpy library. 

````python
import matplotlib.pyplot as plt
````

uint8 bit depth spans from [0,255]

````python
# Create a dark grayscale image (2D array)
gray = np.zeros((200, 200), dtype="uint8")
#gray.imshow(black,cmpa = "gray", vmin = 0 , vmax = 255)

plt.imshow(gray, cmap = plt.cm.gray)
plt.title("grayscale black image")
plt.xticks([]), plt.yticks([])
plt.show()
````

````python
white = 255 * np.ones((200, 200))
white = np.uint8(white)

plt.imshow(white, cmap = plt.cm.gray, vmin=0, vmax=255)
plt.title("grayscale white image")
plt.xticks([]), plt.yticks([])
plt.show()
````

Question 2 Answer:

````python
def check_odd_even(num):
    if num == 0:
        print('0 is neither even nor odd.')
    elif num % 2 == 0:
        print(f'{num} is even.')
    else:
        print(f'{num} is odd.')

print('testing...')
check_odd_even(0)
check_odd_even(1)
check_odd_even(2)
check_odd_even(3)
check_odd_even(4)
check_odd_even(5)
````

Question 3 Answer:

````python
import numpy as np
import time

# Create a large array using Python list comprehension
size = 10**6  # 1 million elements
python_list = [i for i in range(size)]

# Create a large array using numpy
numpy_array = np.arange(size)

# Measure the time taken to sum the elements using Python list
start_time = time.time()
sum_python_list = sum(python_list)
python_time = time.time() - start_time

# Measure the time taken to sum the elements using numpy array
start_time = time.time()
sum_numpy_array = np.sum(numpy_array)
numpy_time = time.time() - start_time

# Print the results
print(f"Sum using Python list: {sum_python_list}, Time taken: {python_time:.6f} seconds")
print(f"Sum using numpy array: {sum_numpy_array}, Time taken: {numpy_time:.6f} seconds")
print(f"numpy is {python_time / numpy_time:.2f} times faster than Python list for this operation.")
````

Question 4 Answer:

````python
# create a 1D array
my_arr = np.arange(10)
print("Initial my_arr: ", my_arr)
arr_slice = my_arr
print("Initial arr_slice: ", arr_slice)

# change the first element of arr_slice
arr_slice[0] = 55

print("my_arr: ", my_arr)
print("arr_slice: ", arr_slice)
````

It is noticed that the my_arr is also changed since arr_slice is referenced to my_arr.

Solution:

````python
# create a 1D array
my_arr = np.arange(10)
print("Initial my_arr: ", my_arr)
arr_slice = np.copy(my_arr)
print("Initial arr_slice: ", arr_slice)

# change the first element of arr_slice
arr_slice[0] = 55

print("my_arr: ", my_arr)
print("arr_slice: ", arr_slice)
````

Question 5 Answer:

````python
import numpy as np
import matplotlib.pyplot as plt

# Dimensions of the image as 200
image_size = 200 

# Create a grid of coordinates
x = np.arange(image_size)
y = np.arange(image_size)
X, Y = np.meshgrid(x, y) 

# Center coordinates of the circle
center_x, center_y = image_size // 2, image_size // 2

# Radius of the circle as 50
radius = 50

# Calculate distances from center for all pixels
distances = np.sqrt((X - center_x)**2 + (Y - center_y)**2)

# Create a black image (zeros background)
image = np.zeros((image_size, image_size), dtype=np.uint8)

# Set pixels within the circle to white (255)
image[distances <= radius] = 255

# Display the image using Matplotlib
plt.imshow(image, cmap='gray')
plt.title('Black Image with White Circular Spot At The Middle')
plt.axis('off')  
plt.show()
````
